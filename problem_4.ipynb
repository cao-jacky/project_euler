```python
def is_palindrome(test_num):
    if str(test_num) == str(test_num)[::-1]:
        return True
```

```python
palindromes = []
for i in range(1000):
    for j in range(1000):
        temp_val = i*j
        # use only 3 digit numbers
        if len(str(i)) == 3 and len(str(j)) == 3:
            if is_palindrome(temp_val):
                # print(f"{i} * {j} = {temp_val} is a palindrome")
                palindromes.append(temp_val)
```

```python
print(max(palindromes))
```
