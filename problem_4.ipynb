```python
def generate_palindromes():
    begin_val = 1
    
    for i in range(1,10):
        if i == 1:
            begin_val = 0
            
        temp_val = str(i) + str(begin_val) + str(i)
        print(temp_val)
```

```python
def is_palindrome(test_num):
    if str(test_num) == str(test_num)[::-1]:
        return True
```

```python
palindromes = []
for i in range(1000):
    for j in range(1000):
        temp_val = i*j
        if len(str(i)) ==3 and len(str(j)) ==3:
            if is_palindrome(temp_val):
                # print(f"{i} * {j} = {temp_val} is a palindrome")
                palindromes.append(temp_val)
```

```python
print(max(palindromes))
```
