```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def list_primes(limit):
    primes_list = []
    for i in range(limit):
        if is_prime(i):
            primes_list.append(i)  
    return primes_list 
```

```python
prime_nums = list_primes(1_000_000)
for prime_val in prime_nums:
    while input_val % prime_val == 0:
        input_val = input_val // prime_val
        print(prime_val)
```

```python
def prime_factors(input_number):
    factors = []
    test_prime = 2
    
    while input_number > 1:
        while input_number % test_prime == 0:
            factors.append(test_prime)
            input_number /= test_prime
        test_prime += 1
        if test_prime*test_prime > input_number:
            if input_number > 1:
                factors.append(int(input_number))
            break
    return factors
```

```python
input_prime_factors = prime_factors(input_val)
print(max(input_prime_factors))
```
