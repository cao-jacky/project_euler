### attempt1

I think this would have brute forced it, given enough time, but I will look for more efficient solutions

```python
input_limit = 1_000_000_000
```

```python
i = 10_000_000
max_divisor = 20
while i < input_limit:
    all_divisible = []
    for j in range(1, max_divisor+1):
        if i % j == 0:
            all_divisible.append(j)
    if len(all_divisible) == max_divisor:
        print(f"{i} is divisible by all numbers from 1 to {max_divisor}")
    i += 1
    
```

```python
def lcm(a, b):
    # finding the least common multiple
    return abs(a*b) // math.gcd(a, b)

max_divisor = 20
result = 1
for i in range(1, max_divisor + 1):
    print(result, i)
    result = lcm(result, i)
```
